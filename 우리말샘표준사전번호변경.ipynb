```python
# 10개의 데이터파일을 합치기
import pandas as pd
Dic=[]
for i in range(1, 11) :
  num=str(i).zfill(2)                     # 두 자릿수 스트링으로 변환
  df=pd.read_excel('/content/drive/MyDrive/KoreanProject/data/우리말샘표준데이터/DATA_'+num+'.xlsx')
  df_drop=df[['우리말샘 표제어', '우리말샘 의미 번호','표준 표제어', '표준 어꺠번호']].dropna(axis=0)   # 필요한 컬럼만 추출+결측값 행 삭제
  Dic.append(df_drop)                     # 리스트로 저장

df_all=pd.concat(Dic, ignore_index=True)  # concat으로 합치기
```

```python
df_all
```

```python
# 실수형 자료를 -.정수형->스트링으로 바꾼다
df_all[['우리말샘 의미 번호', '표준 어꺠번호']]=df_all[['우리말샘 의미 번호', '표준 어꺠번호']].astype(int)
df_all[['우리말샘 의미 번호', '표준 어꺠번호']]=df_all[['우리말샘 의미 번호', '표준 어꺠번호']].astype(str)
```

```python
df_all['우리말샘 의미 번호']='00'+df_all['우리말샘 의미 번호']
df_all['표준 어꺠번호']='0'+df_all['표준 어꺠번호']
df_all['우리말샘 표제어']=df_all['우리말샘 표제어'].replace('[-^]','',regex=True)
df_all['표준 표제어']=df_all['표준 표제어'].replace('[-^]','',regex=True)
df_all['우리말샘표제어']=df_all['우리말샘 표제어']+df_all['우리말샘 의미 번호']
df_all['표준표제어']=df_all['표준 표제어']+df_all['표준 어꺠번호']
df_all.head(10)
```

```python
df_all[['우리말샘표제어',	'표준표제어']].to_excel("/content/drive/MyDrive/KoreanProject/가공data/우리말샘_표준_대조.xlsx", index=False)
```
